# Automatic Red Eye Remover
some time when we click photos through camera at night due to this eyes get bright red color which not look good. In this hole code we are removing the red eye color without losing any pixel texure.

### Code flow
1. Detect Eyes
2. pull the eye image
3. select red area through thresholding
4. dailate
5. fill color in selected area
6. paste the recover image back to original image

```python
import matplotlib.image as Image
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

## Readding Image

```python
red1=Image.imread("red1.jpg")
red1_gray=cv.cvtColor(red1,cv.COLOR_RGB2GRAY)
plt.imshow(red1)
```

## Preparing The classifier

```python
eye_detector=cv.CascadeClassifier("eye_haar.xml")
eyes=eye_detector.detectMultiScale(red1,scaleFactor=1.3, minNeighbors=10, minSize=(50, 50))   
```

## Detecting the Image

```python
copied=red1.copy()

for eye in eyes:
    x,y,w,h=eye
    cv.rectangle(copied,(x,y),(x+w,y+h),(255,0,0),2)
    
plt.figure(figsize=(10,20))
plt.imshow(copied)
```

## Crop Eyes

```python
eyes_a=[]
```

```python


for eye in eyes:
    x,y,w,h=eye
    ey=red1[y:y+h,x:x+w]
    eyes_a.append(ey)
    
f,a=plt.subplots(1,2)
a[0].imshow(eyes_a[0])
a[1].imshow(eyes_a[1])
```

## Spliting in the Rgb format

```python
# Split eye image into 3 channels
eye1=eyes_a[0]

r = eye1[:, :, 0]
g = eye1[:, :, 1]
b = eye1[:, :, 2]


f,a=plt.subplots(1,3,figsize=(12,5))
a[0].imshow(r,cmap='gray')
a[0].set_title("red")
a[1].imshow(g,cmap='gray')
a[1].set_title("green")
a[2].imshow(b,cmap='gray')
a[2].set_title("blue")
```

## Thresholding

```python
#thresholding to detect red area

red_thres=(r>130)
blue_green_thres=(r>cv.add(b,g))
combined_thres=red_thres & blue_green_thres

f,a=plt.subplots(1,3,figsize=(10,5))
a[0].imshow(red_thres,cmap="gray")
a[0].set_title("Red Thresholding")

a[1].imshow(blue_green_thres,cmap="gray")
a[1].set_title("blue_green Thresholding")

a[2].imshow(combined_thres,cmap="gray")
a[2].set_title("Combined Thresholding")
```

## Filling Center dot

```python
def fill_center_dot(mask):
    #making copy of mask in we are going to fill
    flood = mask.copy()
    h, w = flood.shape[:2]
    #cretinf temp mask for the flood fill function
    maskTemp = np.zeros((h+2, w+2), np.uint8)
    
    #floodfill algo
    cv.floodFill(flood, maskTemp, (0, 0), 255)
    #changing white pixel to black and vica-versa
    inverse_flood = cv.bitwise_not(flood)
    #bitwise oring
    return inverse_flood | mask


```

```python
mask=fill_center_dot(combined_thres.astype('uint8'))
mask=cv.dilate(mask,None, iterations=3)
plt.imshow(mask,cmap="gray")
```

## Final Output

```python
mean_bg=cv.add(b,g)/2
mask=mask.astype('bool')

co=r.copy()
np.copyto(co,mean_bg.astype('uint8'),where=mask)

#combined all three layers
final=np.dstack((co,g,b))

plt.imshow(final,cmap="gray")
```

```python
def image_process(eye):
    """
    Image should be in Rgb format
    
    """
    r = eye[:, :, 0]
    g = eye[:, :, 1]
    b = eye[:, :, 2]
    
    red_thres=(r>130)
    blue_green_thres=(r>cv.add(b,g))
    combined_thres=red_thres & blue_green_thres
    
    
    mask=fill_center_dot(combined_thres.astype('uint8'))
    mask=cv.dilate(mask,None, iterations=3)
    
    mean_bg=cv.add(b,g)/2
    mask=mask.astype('bool')
    
    red_layer=r.copy()
    np.copyto(red_layer,mean_bg.astype('uint8'),where=mask)
    
    final=np.dstack((red_layer,g,b)).astype("uint8")
    
    return final
    
```

## Final PipeLine
put all code together

```python
def pipeline(image):
    image_c=image.copy()
    #detect eyes in it
    eye_detector=cv.CascadeClassifier("eye_haar.xml")
    eyes=eye_detector.detectMultiScale(image_c,scaleFactor=1.3, minNeighbors=10, minSize=(50, 50)) 
    print(len(eyes))
    for eye in eyes:
        x,y,w,h=eye
        
        ey=image_c[y:y+h,x:x+w]
        ey_out=image_process(ey)
        
        #recover image paste on the original image
        
        image_c[y:y+h, x:x+w]=ey_out
        
        
    return image_c
    
    
    
```

```python
plt.figure(figsize=(10,20))
plt.imshow(pipeline(red1))
```
